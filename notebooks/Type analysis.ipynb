```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

pd.set_option("display.max_rows", 120)
```

```python
performance = pd.DataFrame.from_csv('/home/haowu4/codes/finer_annotation/results_analysis.tsv', index_col = None)
performance = performance.set_index("type")
```

```python
performance.sort(["corr"], ascending=False)
```

```python
performance_by_rank = performance["corr"].tolist()[::-1]
plt.figure(figsize=(10,10))
plt.plot(performance_by_rank)
plt.axhline(y=50, color='r', linestyle='--')
plt.axvline(x=30, color='r', linestyle='--')

# plt.plot([50] * len(performance_by_rank), "g--")
plt.title("How clean is distance supervision data ")
plt.ylabel("DS accuracy%")
plt.xlabel("Rank")
plt.xticks(np.arange(0,len(performance_by_rank) + 10, 10))
plt.yticks(np.arange(0,110, 10))

plt.show()
```

```python
performance_by_class = pd.DataFrame.from_csv('/home/haowu4/codes/others/figer/aaai/exp.perclass.tsv', index_col = None, sep="\t")
performance_by_class = performance_by_class.sort(["f1", "pred_count"], ascending=[False,True])
# performance_by_class = performance_by_class["." in performance_by_class["label"]]
# performance_by_class = performance_by_class.set_index("label")
performance_by_class = performance_by_class[performance_by_class.apply(lambda x: "." in x['label'], axis=1)]
performance_by_class.index = range(1,len(performance_by_class) + 1)
performance_by_class
```

```python
f1s = performance_by_class["f1"].tolist()
plt.figure(figsize=(10,10))
plt.plot(f1s)
plt.axhline(y=.50, color='r', linestyle='--')
plt.axvline(x=4, color='r', linestyle='--')

# plt.plot([50] * len(performance_by_rank), "g--")
plt.title("Fine Type Performance by rank")
plt.ylabel("F1%")
plt.xlabel("Rank")
plt.xticks(sorted([4] + np.arange(0,len(f1s) + 10, 10).tolist()))
plt.yticks(np.arange(0.0,1.10, 0.10))

plt.show()
```
